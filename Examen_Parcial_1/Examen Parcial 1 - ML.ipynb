## 2. Resultados Mundiales

```python
cols = confirmed_cases.keys()
confirmed = confirmed_cases.loc[:, cols[4]:cols[-1]]
deaths = death_cases.loc[:, cols[4]:cols[-1]]

dates = confirmed.keys()
world_cases = []
total_deaths = []

for i in dates:
    confirmed_sum = confirmed[i].sum()
    death_sum = deaths[i].sum()
    world_cases.append(confirmed_sum)
    total_deaths.append(death_sum)

days_since = np.array([i for i in range(len(dates))]).reshape(-1,1)
world_cases = np.array(world_cases).reshape(-1,1)
total_deaths = np.array(total_deaths).reshape(-1,1)
```

<font color = "red">Describe que es lo que sucede en el bloque de código anterior</font>

**En este bloque de código lo que estamos haciendo es extraer los datos que están apartir de la columna 4 los cuales corresponden a las fechas, que la principal es del 22 de enero del 2020, esto lo hacemos con ayuda de cols que es la variable que tiene almacenadas las llaves de la base de datos. Este paso de localizar las columnas apartir de la columna número cuatro, lo hacemos para el dataframe de casos confirmados y de muertes. Después se crea la variable dates la cual contiene las keys que son todas las fechas, así mismo se crea dos arreglos los cuales nos van a ayudar a guardar el total de casos confirmados en el arreglo de "world_cases" y el total de muertes en el arreglo de "total_deaths". Entonces lo que hacemos en el for es ir sumando los casos confirmados y de muertes, después esos se le adjuntan a sus correspondientes arreglos y al final se le da reshape a los arreglos para que muestren los datos en una columna y que cada dato se muestre en un renglón. days_since muestra desde el día 0 hasta el 606, world_cases muestra todos los casos confirmados en los días mencionados y total_deaths muestra todos las muertes en los días mencionados.**

```python
days_in_future = 10
future_forecast = np.array([i for i in range(len(dates)+days_in_future)]).reshape(-1,1)
adjusted_dates = future_forecast[:-10]
```

```python
start = '1/22/2020'
start_date = datetime.datetime.strptime(start, '%m/%d/%Y')
future_forecast_dates = []
for i in range(len(future_forecast)):
    future_forecast_dates.append((start_date + datetime.timedelta(days=i)).strftime('%m/%d/%Y'))
```

<font color = "red">Describe para que hacemos el código anterior</font>

**Lo que hacemos en esta parte del código, es hacer las predicción de los siguientes 10 días, lo que se hace en la primera parte es definir que los días que queremos predecir son 10 más adelante de la última fecha lo guardamos en la variable 'days_in_future', después en la variable 'future_forecast' hacemos un nuevo array en donde vamos a incluir los diez días mencionados ahora en lugar de ser 606, serán 616 y en la variable 'adjusted_dates' guardamos el arreglos con el número de fechas principales, sin las diez adicionales. En la siguiente parte del código lo que hacemos es darle formato a los días, para que tal cual salgan como fechas, definiendo la fecha principal cómo la que nos indicó el profesor en las instrucciones que es el '1/22/2020' y llegando hasta el '9/29/2021' que son diez días más de la fecha que se tiene en el dataframe, todo esto se hace con el for, en donde en cada iteración se va incrementando el día, pero siempre dandole el formato '%m/%d/&Y'**

Para esta sección vamos a utilizar el conjunto de datos del cáncer de seno o BreastCancer que se encuentra disponible dentro de la librería scikit learn.

Deberán implementar los algoritmos de clasificación que hemos visto en clase y determinar cual de ellos es el que muestra una mejor precisión al momento de realizar la clasificación correspondiente.

El split de datos se realizará con 10%, por lo que deberán mostrar los valores de precisión y la matriz de confusión correspondiente de cada uno de los algoritmos. Deberán explicar la matriz de confusión y cómo afecta a los resultados obtenidos.

Finalmente, deberán realizar un resumen de las precisiones obtenidas y explicar cuál algoritmo y porqué resulto ser más preciso que los otros.

### ORDEN DE ALGORITMOS
* Decision Tree
* Random Forest
* SVM (kernel - lineal)

```python
########## LIBRERÍAS A UTILIZAR ##########
#Se importan la librerias a utilizar
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
########## PREPARAR LA DATA ##########
#Importamos los datos de la misma librería de scikit-learn
dataset = datasets.load_breast_cancer()
#Guardar el dataSet por target y feature
df_bc = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df_bc['type'] = dataset.target
```

```python
dataset
```

```python
dataset.target_names
#Malignant = 0, benign = 1
```

```python
df_bc.head()
```

```python
#Seleccionamos todas las columnas

#Defino los datos correspondientes a las etiquetas
X = df_bc.drop(columns = 'type') 
y = df_bc['type'] #maligno o benigno = cancer o no
```

```python
########## IMPLEMENTACIÓN DE K VECINOS MÁS CERCANOS ##########
#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1)
```

## 1.Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
print("#----------------------------------#")
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_train_pred = dtc.predict(X_train)
print("Matríz de Confusión")
matrix = confusion_matrix(y_train, dtc.predict(X_train), labels=[1,0])
print(matrix)
print("Precisión del modelo")
print(accuracy_score(y_test, dtc.predict(X_test)))
```

## 2. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
print("#----------------------------------#")
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
print("Matríz de Confusión")
matrix = confusion_matrix(y_train, rfc.predict(X_train), labels=[1,0])
print(matrix)
print("Precisión del modelo")
print(accuracy_score(y_test, rfc.predict(X_test)))
```

## 3. SVM

```python
from sklearn.svm import SVC
print("#----------------------------------#")
svc = SVC() #lineal
svc.fit(X_train, y_train)
print("Matríz de Confusión")
matrix = confusion_matrix(y_train, svc.predict(X_train), labels=[1,0])
print(matrix)
print("Precisión del modelo")
print(accuracy_score(y_test, svc.predict(X_test)))
```

### La matríz de confusión está conformada de la siguiente manera:
    
    
[ ............... ]            [ Maligno ]                              [ Benigno ] <br>
<br>
[ Maligno  ]  [ N verdaderos positivos ]              [ N falsos negativos  (Error Tipo 2) ] <br>
<br>
 [ Benigno  ]  [ N falsos positivos (Error Tipo 1)]    [ N  verdaderos negativos]

En cuanto a los algoritmos utilizados para la demostración de clasificación, consideramos que el Random Forest es uno de los más adaptables, ya que, al ser un tema, el cáncer de seno, que ocupan muchas variables, es importante tener en cuenta cada una de ellas. 

El Random Forest se encarga de ello, por ende, ocupa más tiempo en ejercer el modelo de predicción adecuado.
Por otra parte, el algoritmo de Decision Tree es un poco más eficaz al lidiar con el número de datos que se están evaluando en el dataset. Sin embargo, si intentamos observar un caso de la vida real, necesitaríamos confiarnos de aquel que tome en cuanto grandes cantidades de variables, contrario a lo que se ejecuta con el Decision Tree.

Por último, el algoritmo de SVM, fue el modelo menos preciso de todos los algoritmos que ocupamos, tomando en cuenta que es un modelo que se basa en distancias entre un hiper plano y el punto más cercano a cierta variable.

Sin duda alguna, cualquiera de estos algoritmos es útil al momento de clasificar un dataset, ya que esta enfermedad se diagnostica mediante variables con alto grado de certeza.
