```python
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split, cross_validate, cross_val_score
```

```python
features = heart_df.drop(columns = 'HeartDisease').keys()
features
```

Cross validation con kFolds aleatorio

```python
dfs = []
kf = KFold(n_splits=5, shuffle=True, random_state=123)
i = 1

for train_index, test_index in kf.split(heart_df):
    X_train = heart_df.iloc[train_index].loc[:, features]
    X_test = heart_df.iloc[test_index].loc[:, features]
    y_train = heart_df.iloc[train_index].loc[:,'HeartDisease']
    y_test = heart_df.loc[test_index].loc[:,'HeartDisease']
    
    model.fit(X_train, y_train) #Training the model
    y_pred = model.predict(X_test)
    print(f"Accuracy for the fold no. {i} on the test set: {accuracy_score(y_test, y_pred)}")
    
    # how many occurances appear in the train set
    s_train = heart_df.iloc[train_index].loc[:,'HeartDisease'].value_counts()
    s_train.name = f"train {i}"
    s_test = heart_df.iloc[test_index].loc[:,'HeartDisease'].value_counts()
    s_test.name = f"test {i}"
    df = pd.concat([s_train, s_test], axis=1, sort=False)
    df["|"] = "|"
    dfs.append(df)
    
    i += 1
```
